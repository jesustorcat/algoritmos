```python
# Creando el Nodo

class Nodo:
    
    def __init__(self, nombre = None, rut = None, hijo_izquierdo = None, hijo_derecho = None):
        
        self.nombre = nombre
        self.rut = rut
        self.hijo_izquierdo = hijo_izquierdo
        self.hijo_derecho = hijo_derecho
        
    def __str__(self):
        return ' %s %s' %(self.nombre, self.rut)
```

```python
# Creando el Arbol

class Arbol_Binario:
    
    def __init__(self, raiz = None):
        self.raiz = raiz
        
    def agregar(self, elemento):
        if self.raiz == None:
            self.raiz = elemento
        else:
            auxiliar = self.raiz
            padre = None
            while auxiliar != None:
                padre = auxiliar
                if int(elemento.rut) >= int(auxiliar.rut):
                    auxiliar = auxiliar.hijo_derecho
                else:
                    auxiliar = auxiliar.hijo_izquierdo
            if int(elemento.rut) >= int(padre.rut):
                padre.hijo_derecho = elemento
            else:
                padre.hijo_izquierdo = elemento
                
    def pre_orden(self, elemento):
        if elemento != None:
            print(elemento)
            self.pre_orden(elemento.hijo_izquierdo)
            self.pre_orden(elemento.hijo_derecho)
            
    def post_orden(self, elemento):
        if elemento != None:
            self.post_orden(elemento.hijo_izquierdo)
            self.post_orden(elemento.hijo_derecho)
            print(elemento)
            
    def in_orden(self, elemento):
        if elemento != None:
            self.post_orden(elemento.hijo_izquierdo)
            print(elemento)
            self.post_orden(elemento.hijo_derecho)
    
    def obtener_raiz(self):
        return self.raiz
```

```python
# Ejecutando programa arbol

if __name__ == '__main__':
    arbol = Arbol_Binario()
    bandera = True
    
    while(bandera):
        print('______________Menu__________________\n'+
             '1. Agregar.\n'+
             '2. Pre-Orden.\n'+
             '3. Post-Orden\n'+
             '4. In-Orden\n'+
             '5. Salir.')
        
        opcion = input('Ingrese la opción: ')
        
        if opcion == '1':
            Nombre = input('Ingrese el nombre: ')
            Rut = input('Ingrese el R.U.T: ')
            nodo = Nodo(Nombre, Rut)
            arbol.agregar(nodo)
        elif opcion == '2':
            print('Imprimiendo por Pre-Orden')
            arbol.pre_orden(arbol.obtener_raiz())
        elif opcion == '3':
            print('Imprimiendo por Post-Orden')
            arbol.post_orden(arbol.obtener_raiz())
        elif opcion == '4':
            print('Imprimiendo por inorden')
            arbol.in_orden(arbol.obtener_raiz())
        elif opcion == '5':
            bandera = False
    
```
